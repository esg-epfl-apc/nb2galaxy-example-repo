```python
import numpy as np
import matplotlib.pyplot as pt
```

```python
min_ = -100 #http://odahub.io/ontology#float
max_ = 100 #http://odahub.io/ontology#float
power = 3 #http://odahub.io/ontology#integer
```

```python
x = np.linspace(min_, max_, 1001)
pt.figure(figsize=(5, 4))
im = pt.plot(x, x ** power)
pt.xlabel("x")
pt.ylabel(f"$y^{power}$")
```
