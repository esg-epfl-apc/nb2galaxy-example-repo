```python
import numpy as np
from numpy.polynomial import Polynomial
import matplotlib.pyplot as plt
```

```python
coeff_f = 'dataset.csv' # oda:POSIXPath
min_ = -2 # oda:Float; oda:label "Left bound of x-axis"
max_ = 4 # oda:Float; oda:label "Right bound of x-axis"
ymin = None # oda:Float, oda:optional; oda:label "Lower bound of y-axis" 
ymax = None # oda:Float, oda:optional; oda:label "Upper bound of y-axis"
```

```python
coeff = np.genfromtxt(coeff_f, delimiter=',')

p = Polynomial(coeff)
```

```python
x = np.linspace(min_, max_, 1000)
plt.figure(figsize=(5, 4))
im = plt.plot(x, p(x), label=f"{p}")
plt.xlabel("x")
plt.legend()
plt.ylim(ymin, ymax)

plt.savefig('.plot.png')
```
