```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
min_ = -10
max_ = 10
power = 3
```

```python
x = np.linspace(min_, max_, 1001)
plt.figure(figsize=(5, 4))
im = plt.plot(x, x ** power)
plt.xlabel("x")
plt.ylabel(f"$x^{power}$")

plt.savefig('.plot.png')
```
