```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
power = 3 # oda:Integer
min_ = -10 # oda:Float; oda:label "Left bound of x-axis"
max_ = 10 # oda:Float; oda:label "Right bound of x-axis"
ymin = None # oda:Float, oda:optional; oda:label "Lower bound of y-axis" 
ymax = None # oda:Float, oda:optional; oda:label "Upper bound of y-axis"
```

```python
x = np.linspace(min_, max_, 1000)
plt.figure(figsize=(5, 4))
im = plt.plot(x, x ** power)
plt.xlabel("x")
plt.ylabel(f"$x^{{{power}}}$")
plt.ylim(ymin, ymax)

plt.savefig('.plot.png')
```
